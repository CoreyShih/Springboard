# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

### B. Provide support for your arguments and your own recommendations with a statistically sound analysis.

First let's visualize the excess readmission rate distributions for hospitals with < 100 discharges and hospitals with > 1000 discharges.

```python
small = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] < 100]
large = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] > 1000]
```

```python
plt.figure(figsize=(8, 5))
plt.subplot(1, 2, 1)
small['Excess Readmission Ratio'].hist(bins=20, normed=True)
plt.title('< 100 discharges')
plt.xlabel('Excess Readmission Ratio')
plt.subplot(1, 2, 2)
large['Excess Readmission Ratio'].hist(bins=20, normed=True)
plt.title('> 1000 discharges')
plt.xlabel('Excess Readmission Ratio')
plt.show()
```

```python
plt.figure(figsize=(5, 5))
ax = plt.subplot(1, 2, 1)
small.boxplot('Excess Readmission Ratio', return_type='axes')
plt.title('< 100 discharges')
plt.subplot(1, 2, 2, sharey=ax)
large.boxplot('Excess Readmission Ratio', return_type='axes')
plt.title('> 1000 discharges')
plt.show()
```

One thing we can immediately notice from these plots is that there are a lot more high outliers for small facilities than large ones, which could be driving the mean excess readmission ratio up.

For the hypothesis test, we will use a 2 sample t test for the difference of means between small and large facilities. The null hypothesis is that there is no difference in the mean excess readmission ratio between small and large hospitals, and the alternative hypothesis is that small hospitals have a higher mean excess readmission ratio than large hospitals. We will use bootsrapping to simulate resampling the data many times and $\alpha$ = .01.

```python
def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""
    reps = np.array([func(np.random.choice(data, size=len(data))) for _ in range(size)])
    return reps
```

```python
# Compute total mean for small and large hospitals
mean_total = np.mean(pd.concat([small, large])['Excess Readmission Ratio'])

# Shift means of samples
small_shifted = small['Excess Readmission Ratio'] - np.mean(small['Excess Readmission Ratio']) + mean_total
large_shifted = large['Excess Readmission Ratio'] - np.mean(large['Excess Readmission Ratio']) + mean_total
small_shifted = small_shifted.dropna()
large_shifted = large_shifted.dropna()

# Draw 100,000 bootstrap replicates from shifted samples
small_reps = draw_bs_reps(small_shifted, np.mean, 100000)
large_reps = draw_bs_reps(large_shifted, np.mean, 100000)

# Find replicates of the difference of means
bs_reps = small_reps - large_reps

# Calculate p-value
p = np.sum(bs_reps >= np.mean(small['Excess Readmission Ratio']) - np.mean(large['Excess Readmission Ratio'])) / len(bs_reps)
print('p-value: {}'.format(p))
```

The p-value is so small that not 1 in 100,000 bootsrap replicates displayed a difference in means as large as was observed. We can repeat the analysis with frequentist statistics to find the p-value.

```python
from scipy import stats
```

```python
s = small['Excess Readmission Ratio'].dropna()
l = large['Excess Readmission Ratio'].dropna()
v = np.var(pd.concat([s, l]))

# Compute t-score
t = (np.mean(s) - np.mean(l)) / np.sqrt(v * (1/float(len(s)) + 1/float(len(l))))

# Compute p-value
p = stats.t.sf(np.abs(t), len(s) + len(l) - 2)
print('p-value: {}'.format(p))
```

The p-value is 2.9e-22, which is much smaller than $\alpha$ = .01, so we reject the null hypothesis and conclude that hospitals with < 100 discharges have a higher mean excess readmission ratio than hospitals with > 1000 discharges. With a p-value this low, we can practically say that there is likely a large difference between small and large hospitals, and so the recommendations should be followed.

Regarding the scatterplot, one of the advantages of the plot is that is shows every single data point; however, the plot is extremely cluttered due to the large number of data points with low numbers of discharges and excess readmission ratios around 1. The important information to be conveyed here is that smaller hospitals have higher excess readmission ratios, which can be done with a series of boxplots.

```python
med = clean_hospital_read_df[(
    clean_hospital_read_df['Number of Discharges'] >= 100) & (clean_hospital_read_df['Number of Discharges'] <= 1000)]

plt.figure(figsize=(8, 8))
ax = plt.subplot(1, 3, 1)
small.boxplot('Excess Readmission Ratio', return_type='axes')
plt.title('< 100 discharges')
ax = plt.subplot(1, 3, 2, sharey=ax)
med.boxplot('Excess Readmission Ratio', return_type='axes')
plt.title('100-1000 discharges')
plt.subplot(1, 3, 3, sharey=ax)
large.boxplot('Excess Readmission Ratio', return_type='axes')
plt.title('> 1000 discharges')
ax.set_ylim(0.6, 1.4)
plt.tight_layout()
plt.show()
```
