```python
import pandas as pd

df = pd.read_csv('data/human_body_temperature.csv')
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

np.random.seed(1)
```

### 1. Is the distribution of body temperatures normal?

First, let's compare the a histogram of the data to a normal distribution using the sample mean and standard deviation.

```python
mean, std = stats.norm.fit(df.temperature)

%matplotlib inline

plt.hist(df.temperature, bins=20, normed=True, alpha=0.5)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax)
y = stats.norm.pdf(x, mean, std)
plt.plot(x, y, color='k')
plt.xlabel('body temperature (F)')
plt.ylabel('pdf')
plt.show()
```

This looks like the data could come from a normal distribution. Let's also plot the ECDF of the data:

```python
def ecdf(data):
    x = np.sort(data)
    y = np.arange(1, len(x) + 1) / float(len(x))
    return x, y
```

```python
x_data, y_data = ecdf(df.temperature)
x_norm, y_norm = ecdf(np.random.normal(mean, std, 1000))

plt.plot(x_data, y_data, marker='.', linestyle='none', alpha=0.5)
plt.plot(x_norm, y_norm, marker='.', linestyle='none', alpha=0.5)
plt.xlabel('body temperature (F)')
plt.ylabel('ecdf')
plt.legend(['data', 'normal'])
plt.show()
```

The ECDF of the data looks to closely match the ECDF of a normal distribution. As a final test, let's use a normality test to test the null hypothesis that the data comes from a normal distribution with an alpha of 0.05.

```python
_, p = stats.normaltest(df.temperature)
print('p = {}'.format(p))
```

We find a rather large p-value of 0.26. This is larger than our alpha of 0.05, so we cannot reject the null hypothesis that the sample comes from a normal distribution. This result, combined with the graphical similarities between the ECDF and PDF of the data to those of normal distributions, suggests that the distribution of body temperatures is indeed normal.

### Is the true population mean really 98.6 degrees F?

Let's first see what our sample mean is.

```python
print(mean)
```

To test the null hypothesis that the true population mean really is 98.6 degrees F, we will translate the mean of our sample to be equal to 98.6 and braw bootstrap replicates. We can then calculate the approximate probability of measuring a sample mean as low as our observed mean of 98.25 assuming the null hypothesis is true. We will use an alpha of 0.05.

```python
def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""
    reps = np.array([func(np.random.choice(data, size=len(data))) for _ in range(size)])
    return reps
```

```python
# Translate the mean of our sample to 98.6
translated_temps = df.temperature - mean + 98.6

# Draw 100000 bootsrap replicates of the translated sample
bs_reps = draw_bs_reps(translated_temps, np.mean, 100000)

# Calculate the p-value
p = np.sum(bs_reps <= mean) / float(len(bs_reps))
print('p = {}'.format(p))
```

We obtain a p-value of 0, which means that in none of the 100000 bootstrap replicates was the mean body temperature as low as 98.25 degrees F. The actual p-value is very small, on the order of 1e-5. This p-value is much smaller than our alpha of 0.05, so we reject the null hypothesis and conclude that the true population mean is not 98.6 degrees F, but lower.

Next we will use statistical testing to arrive at the same conclusion. We will use a 1 sample test since we are comparing a single sample to a statistic (98.6 degrees F). A t test is appropriate here because we do not know the population standard deviation and must estimate it from the sample standard deviation.

```python
alpha = 0.05
t = (mean - 98.6) / (std / np.sqrt(df.shape[0]))
p = stats.t.sf(np.abs(t), df.shape[0] - 1)
print('t-score: {}'.format(t))
print('p-value: {}'.format(p))
```

We see that the p-value is of much less than 0.05, so we reject the null hypothesis and conclude that the true population mean is less than 98.6 degrees F.

Next, we'll try the same calculation iwht a z-test instead.

```python
z = (mean - 98.6) / (std / np.sqrt(df.shape[0]))
p = stats.norm.sf(np.abs(z))
print('z-score: {}'.format(z))
print('p-value: {}'.format(p))
```

Our p-value here is smaller than that of the t-test, since the t distribution has bigger tails than the normal distribution. The conclusion is the same, however; the p-value is much less than 0.05, so we reject the null hypothesis and conclude that the true population mean is less than 98.6 degrees F.

### 4. Draw a small sample of size 10 from the data and repeat both frequentist tests.

```python
sample = np.random.choice(df.temperature, size=10)
```

```python
t = (np.mean(sample) - 98.6) / (np.std(sample) / np.sqrt(len(sample)))
p = stats.t.sf(np.abs(t), len(sample) - 1)
print('t-score: {}'.format(t))
print('p-value: {}'.format(p))
```

```python
z = (np.mean(sample) - 98.6) / (np.std(sample) / np.sqrt(len(sample)))
p = stats.norm.sf(np.abs(z))
print('z-score: {}'.format(z))
print('p-value: {}'.format(p))
```

Here, we see the differences in the t and z tests for small sample sizes. The correct test to use in this case is the t test, since we have a small sample size and are estimating the population standard deviation. Both tests fail to reject the null hypothesis, but the p-value for the z test is smaller. For small sample sizes, the difference between the 2 tests becomes more pronounced, which is why it is important to use the correct one.

### 5. At what temperature should we consider someone's temperature to be "abnormal"?

To start, we can take bootsrap replicates of the data to simulate resampling the data and construct an interval in which 99% of body temperatures lie within.

```python
# Draw 100000 bootsrap replicates
bs_reps = [np.random.choice(df.temperature, size=len(df.temperature)) for _ in range(100000)]
conf_int = np.percentile(np.concatenate(bs_reps), [0.5, 99.5])
print(conf_int)
```

We can consider any body temperature which lies outside this interval to be abnormal. We can do the same thing with frequentist statistics using t statistics.

```python
t = stats.t.ppf(0.995, len(df.temperature) - 1)
margin = t * std
conf_int = [mean - margin, mean + margin]
print('margin of error: {}'.format(margin))
print('confidence interval: {}'.format(conf_int))
```

Assuming the population is normally distributed, 99% of body temperatures should fall within this interval, and anything outside of it can be considered abnormal.

### 6. Is there a significant difference between males and females in normal temperature?

For this question, we will use a two sample t test to test the null hypothesis that there is no difference between the mean body temperatures between males and females. We will use an alpha of 0.05. We will need to calculate the difference in mean temperatures between males and females in the data set and the variances for both sexes as well. Then we will calculate the probability of observing a difference in means as large as what we found in the data set assuming there is no difference between sexes. We will also assume equal variances for males and females.

```python
males = df[df.gender == 'M']
females = df[df.gender =='F']
mean_m = np.mean(males.temperature)
mean_f = np.mean(females.temperature)
mean_diff = mean_m - mean_f
std_diff = np.sqrt(np.var(df.temperature)/len(males) + np.var(df.temperature)/len(females))
print('difference in mean body temperature between males and females: {}'.format(mean_diff))
```

```python
t = mean_diff / std_diff
p = stats.t.sf(np.abs(t), len(males) + len(females) - 2)
print('p = {}'.format(p))
```

We find a p-value of 0.013, which is less than our alpha of 0.05, so we reject the null hypothesis and conclude that men have lower mean body temperature than women.

To summarize, we have so far concluded in this assignment that the true mean body temperature is lower than Carl Wunderlich's reported value of 98.6 degrees F, and that the mean body temperature of females is higher than that of males. Putting these two together seems to suggest that in calculating the mean body temperature, Carl Wunderlich may have sampled from more women than men, which caused his reported value to be higher than the true population mean.
